```python
def gernerate_df(playlist_link):
    playlist_URI = playlist_link.split("/")[-1].split("?")[0]
    track_uris = [x["track"]["uri"] for x in sp.playlist_tracks(playlist_URI,offset=100)["items"]]
    df = pd.DataFrame()
    for track in sp.playlist_tracks(playlist_URI)["items"]:
        #URI
        track_uri = track["track"]["uri"]
        dic = sp.audio_features(track_uri)[0]

        #Track name
        track_name = track["track"]["name"]
        dic['track_name'] = track_name

        #Main Artist
        artist_uri = track["track"]["artists"][0]["uri"]
        artist_info = sp.artist(artist_uri)

        #Name, popularity, genre
        artist_name = track["track"]["artists"][0]["name"]
        artist_pop = artist_info["popularity"]
        artist_genres = artist_info["genres"]
        dic['artist_name'] = artist_name
        dic['artist_popularity'] = artist_pop

        #Album
        album = track["track"]["album"]["name"]

        #Popularity of the track
        track_pop = track["track"]["popularity"]
        dic['track_popularity'] = track_pop

        df = df.append(dic, ignore_index=True)

        
    return df
```

```python
df1 = gernerate_df("https://open.spotify.com/playlist/3IwKznBqqtET82TWTPicV2?si=7c30cad409774a97")
df2 = gernerate_df("https://open.spotify.com/playlist/5a7c7GrdcmI9vXmxR3dtWQ?si=f1b10bb09bb048be")
df3 = gernerate_df("https://open.spotify.com/playlist/0xbst7aj9t5S8f6gXTrWKZ?si=c682aecff83b48d9")
df4 = gernerate_df("https://open.spotify.com/playlist/43pfR7Ca4ylSsEUMHzevyf?si=054b750aeca7455e")
df5 = gernerate_df("https://open.spotify.com/playlist/4Ja0KABNq9pzihU3uKmZuM?si=5527c88de5eb42ce")
df6 = gernerate_df("https://open.spotify.com/playlist/3gXM9dIGJgiLmJsMwQu7fU?si=80b855f428e84835")
df7 = gernerate_df("https://open.spotify.com/playlist/2Co7ei0MccUmgtRjYh1F2w?si=61814ff4305a434e")
df8 = gernerate_df("https://open.spotify.com/playlist/78AqEPNLZuGlq4z0TGfWDG?si=349bd772f6e444dd")
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8])
df.shape
```
