```python
#Imports
import pandas as pd
import altair as alt
alt.__version__
```

## Drought Data
Obtained from: https://droughtmonitor.unl.edu/DmData/DataDownload/DSCI.aspx
- Selected states and calendar year 2016
- DSCI: Drought Severity and Coverage Index
Converts drought levels to a single value
Range is 0-500 with zero meaning none of the area is abnormally dry or in drought, a score of 500 means that all areas is in exceptional drought
- Recommendation is to average the value over a period of time if attempting to get a single value


```python
drought_df = pd.read_csv('drought_monitor_export_20160101_20161231.csv')  
```

```python
drought_df['Name'].value_counts()
```

```python
drought_df.shape
```

```python
# Aggregate by state average
mean_state_drought = pd.DataFrame({'mean_DSCI': drought_df.groupby('Name')['DSCI'].mean().round(decimals = 2)}).reset_index()
mean_state_drought
```

```python
#Bar chart of states: 
bars = alt.Chart(mean_state_drought, title= 'Mean Drought Severity by State for 2016: Zero = no drought, 500 = exceptional drought')\
.mark_bar().encode(
    x='mean_DSCI:Q',
    y= 'Name:O',
    color= alt.Color('mean_DSCI', legend=alt.Legend(title= 'Drought Severity'),\
                     scale=alt.Scale(scheme='turbo'))
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text= 'mean_DSCI:Q'
)

(bars + text).properties(height=900)
```
